##Python

```python
#@title
import graphviz
from graphviz import Digraph

py = Digraph('teaching order dependencies', 
             engine='dot')
py.node_attr.update(fillcolor='blue:cyan', fontcolor='white',
        style='filled', gradientangle='270')

py.node('py intro',label="sequence")
py.edge('input→process→output','process')

with py.subgraph(name='cluster process') as pro:
  pro.attr(label='Process', fontcolor='black')
  pro.node('process', shape='point')
  pro.edge('algorithms','algorithm vs program')
  pro.edge('fetch → decode → execute','tracing')

py.node('use procedures')

py.node('arithmetic operators','arithmetic operators:\n +, -, *, /')
py.node('comparison operators','comparison operators:\n<, >, <=, >=, ==, !=')
py.node('boolean operators','boolean operators: and, or, not')
py.edge('integer','arithmetic operators')
py.edge('integer','comparison operators')
py.edge('comparison operators','boolean operators')
py.edge('boolean operators','selection')

py.edge('py intro','use procedures')
py.edge('py intro','structured')
py.edge('use procedures', 'write procedures')
py.edge('structured','write procedures')
py.edge('structured','iteration')
py.edge('structured','selection')
py.edge('write procedures','write procedures\n with args')
py.edge('basic types','write procedures\n with args')
py.edge('basic types','collections')
py.node('collections',label='collection\ndata types')
py.edge('collections','iteration')
with py.subgraph(name='cluster data types') as dt:
  dt.attr(label='Basic data types', fontcolor='black')
  dt.node('basic types', shape='point')
  dt.node('strings')
  dt.node('integer')
  dt.node('float')


display(py)
```

## Web design

```python
#@title
import graphviz
from graphviz import Digraph

web = Digraph('teaching order dependencies', engine='dot')

web.node_attr.update(fillcolor='purple:pink', fontcolor='white',
           style='filled', gradientangle='270')

web.edge('markdown', 'basic html')
web.edge('basic html', 'CSS style')
web.edge('basic html', 'markdown→HTML')
with web.subgraph(name='cluster basic html') as html:
  html.attr(label='Basic HTML', fontcolor='black')
  html.node('basic html', shape='point')
  html.node('headings')
  html.node('paragraphs')
  html.node('bullet lists')
  html.node('images')
  html.node('links')
  html.node('quotes')
with web.subgraph(name='cluster html structure') as html_st:
  html_st.attr(label='HTML Structure', fontcolor='black')
  html_st.node('html structure', shape='point')
  html_st.node('article')
  html_st.node('aside')
  html_st.node('header')
  html_st.node('footer')

web.edge('basic html','html structure')
web.edge('html structure','CSS structure')
web.edge('CSS style','CSS structure')
   
display(web)
```

## Teaching: Programming and web-design — interaction

```python
#@title
g = Digraph('teaching order dependencies', 
            engine='dot')
g.attr(splines='line', nodesep='1')

g.subgraph(web)
g.subgraph(py)
g.edge('colab intro', 'py intro')
g.edge('colab intro', 'markdown')
g.edge('basic html','structured')
g.edge('use procedures','markdown→HTML')
g.edge('strings','markdown→HTML')
g.edge('input→process→output','markdown→HTML')

display(g)
```

In the above graph we see that an early part of python `structured` depends on `Basic HTML` (or at least could benafit from what is learnt there). The dependency the other way is small `markdown→HTML` depends on being able to `use procedures`. However `markdown→HTML` is at the end of web-design (nothing depends on it).

Therefore we should teach web-design first, and move `markdown→HTML` to the end of the programming in python unit.

If we could have a simple command line program to do the markdown→HTML, then we could use it much earlier. However while possible on MS-Windows, it is very klunky. Not an easy thing to do, and certenly not easy to teach a class of 20. If we had a Unix system (such as Gnu/Linux), then this would be easy.
